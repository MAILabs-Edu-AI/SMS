```python
import numpy as np
import matplotlib.pylab as plt
lyamda = 0.5
mu = 1
mu1 = 0.8*mu
mu2 = 0.2*mu
k = 5 #Число каналов +1

a = 0
b = 1
eps = 0.001
delta = 0.0001

def factorial(n):
    f = 1
    i = 1
    while i <= n:
        f *= i
        i += 1
    return f

# Вычисление вероятности отказа по формуле Эрланга
def p_otk(u, mu):
    r = lyamda * u / mu
    sum = 0
    for i in range (k):
        sum += pow(r, i) / factorial(i)
    p0_e = 1 / sum
    return pow(r,3) / factorial(3) * p0_e

# Метод дихотомии
while (abs(a-b)>=eps):
    u1 = (a + b - delta) / 2
    u2 = a + b - u1
    f1 = p_otk(u1, mu1) * u1 * lyamda + p_otk(1 - u1, mu2) * (1 - u1) * lyamda
    f2 = p_otk(u2, mu1) * u2 * lyamda + p_otk(1 - u2, mu2) * (1 - u2) * lyamda
    if (f1 < f2):
        b=u2
    else:
        a=u1
u = (a + b) / 2
print("u1 = {:.4f}".format(u))
print("u2 = {:.4f}".format(1 - u))
print("J'(min) = {:.4f}".format(p_otk(u, mu1) * u * lyamda + p_otk(1 - u, mu2)\
                                * (1 - u) * lyamda))

#Построение графика
x = np.arange(0,1,0.001)
y =  (p_otk(x, mu1)*x*lyamda + p_otk(1-x, mu2)*(1-x)*lyamda)
plt.title("Зависимости: J от x") # заголовок
plt.xlabel("x")         # ось абсцисс
plt.ylabel("J")    # ось ординат
plt.grid()              # включение отображение сетки
plt.plot(x, y, label = "J")# построение графика
plt.legend() #добавим легенду
plt.show() #показ графика

plt.title("Зависимости: J от x") # заголовок
plt.xlabel("x")         # ось абсцисс
plt.ylabel("J")    # ось ординат
plt.grid()              # включение отображение сетки
plt.plot(x, y, label = "J")# построение графика
plt.legend() #добавим легенду
plt.ylim(0, 0.01)
plt.show() #показ графика
```
